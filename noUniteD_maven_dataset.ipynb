```python
datasets_path = './datasets/'

# where the dataset will be downloaded
maven_dataset_path = os.path.join(datasets_path, 'maven/')
maven_train_path = os.path.join(maven_dataset_path, "train.jsonl")
maven_valid_path = os.path.join(maven_dataset_path, "valid.jsonl")
maven_test_path = os.path.join(maven_dataset_path, "test.jsonl")

amuse_url = 'http://127.0.0.1:3002/api/model' # put here your endpoint for AMuSE-WSD!
invero_url = 'http://127.0.0.1:3003/api/model' # put here your endpoint for invero-xl!
```

```python
# generating the list of sentences from maven:
import json

intermediate_dataset = []
with open(maven_train_path, 'r') as json_file:
    maven_train = list([json.loads(json_line) for json_line in json_file])
for el in maven_train:
    for i, sentence_tokens in enumerate(el['content']):
        intermediate_dataset.append(sentence_tokens['sentence'])

print(len(intermediate_dataset))
intermediate_dataset[:6]
```

```python
amuse_request(["Marco is running a marathon.", "The eating of Marco is very loud."], amuse_url=amuse_url)
```

```python
invero_request(["Marco is running a marathon.", "The eating of Marco is very loud."], invero_url=invero_url)
```

```python
# parameters:
sentences_testing_dataset = ["Marco is running a marathon.", "The eating of Marco is very loud."] # this is for testing 
sentences_dataset = intermediate_dataset # a list of sentences, the starting dataset
unambiguous_candidates_path = './va_resources/candidates_unambiguous.json' # path to the unambiguous candidates to be used to create the nominal part
chunk_size = 16 # number of sentences to query amuse and invero
window_span_error = 3 # the displacement between invero and amuse tokenization indices. The greater, the less are the incorrelations errors
lang = "EN"

noUniteD_srl_result, nominal_event_count = create_nounited_dataset(
    sentences_list = sentences_dataset,
    unambiguous_candidates_path = unambiguous_candidates_path,
    amuse_url = amuse_url, invero_url = invero_url,
    chunk_size = chunk_size, window_span_error = window_span_error, lang = lang
)
```

```python
nominal_event_count.most_common()
```

```python
select_sample = 0
print([(i,w) for i,w in enumerate(noUniteD_srl_result[select_sample]['words']) ], '\n')
print([(i,w) for i,w in enumerate(noUniteD_srl_result[select_sample]['predicates']) ], '\n')
print(noUniteD_srl_result[select_sample]['roles'], '\n')
print([(i,w) for i,w in enumerate(noUniteD_srl_result[select_sample]['lemmas']) ], '\n')
print([(i,w) for i,w in enumerate(noUniteD_srl_result[select_sample]['pos_tags']) ], '\n')
print(noUniteD_srl_result[select_sample]['num_v'], noUniteD_srl_result[select_sample]['num_n'])
```

```python
save_nounited_dataset(
    noUniteD_srl_result = noUniteD_srl_result, 
    dir_path = './datasets/maven_nounited_srl/', lang = 'EN', 
    train_ratio = 0.8, num_dataset_divisions = 2, shuffle = False
)
```

```python
import json

with open('./datasets/united-srl/EN/train.json', 'r') as json_file:
    united_train = json.load(json_file)
with open('./datasets/united-srl/EN/dev.json', 'r') as json_file:
    united_dev = json.load(json_file)
```

```python
with open('./datasets/maven_nounited_srl/EN/train.json', 'r') as json_file:
    nounited_train = json.load(json_file)
with open('./datasets/maven_nounited_srl/EN/dev.json', 'r') as json_file:
    nounited_dev = json.load(json_file)
```

```python
preds_in_united_train = set([p for v in united_train.values() for p in v['predicates']])
preds_in_united_dev = set([p for v in united_dev.values() for p in v['predicates']])
preds_in_united_all = set( list(preds_in_united_train) + list(preds_in_united_dev) )

len(preds_in_united_train), len(preds_in_united_dev), len(preds_in_united_all)
```
